```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
symbols = [
    "AAPL",  # Apple
    "MSFT",  # Microsoft
    "GOOGL", # Alphabet (Google)
    "AMZN",  # Amazon
    "TSLA",  # Tesla
    "NFLX",  # Netflix
    "PYPL",  # PayPal
    "ADBE",  # Adobe
    "V",     # Visa
    "NVDA",  # Nvidia
]
```

```python
tickers = symbols
data = yf.download(tickers, start='2023-01-01', end='2024-01-01')
adj_close = data['Adj Close']
returns = adj_close.pct_change().dropna()
returns.to_csv('historical_returns.csv')
```
